```
import numpy as np
```

```
#x_m=entradas do treinamento
Xm = np.array([0.5, 0.1])
Wkp= np.array([0.1, 0.2])#pesos_das_entradas
b = 0.3 #bias 
```

```

Uk= np.dot(Wkp, Xm) + b #uk=entrada_para_ativacao 
```

```
#Função de ativação (degrau)
def f(Uk): #função degrau
    if(Uk>=0):
        return 1
    else:
        return 0
```

```
Yk= f(Uk) #Yk = saída do neurônio
```

```
print(f'Valor de saída do neurônio: {Yk}')
```
